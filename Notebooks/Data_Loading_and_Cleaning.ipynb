```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)
```

```python
conflict_df = pd.read_csv('Conflict.csv')

print("Shape :", conflict_df.shape)

conflict_df.head()
```

```python
military_df = pd.read_csv(
    'Military_Spending.csv',
    skiprows=4
)

print("Shape :", military_df.shape)

military_df.head()
```

```python
inflation_df = pd.read_csv(
    'Inflation.csv',
    skiprows=4
)

print("Shape :", inflation_df.shape)

inflation_df.head()
```

```python
oil_df = pd.read_csv('Oil_Data.csv')

print("Shape :", oil_df.shape)

oil_df.head()
```

```python
conflict_df.info()
```

```python
conflict_df.isnull().sum().sort_values(ascending=False)
```

```python
conflict_df.columns.tolist()
```

```python
conflict_df = conflict_df.drop_duplicates()

print("Shape After Removing Duplicates")

conflict_df.shape
```

```python
military_df.columns.tolist()
```

```python
keep_cols = ['Country Name','Country Code']

for year in range(2000,2025):
    keep_cols.append(str(year))

military_df = military_df[keep_cols]

military_df.head()
```

```python
military_df.isnull().sum().sort_values(
    ascending=False
).head(20)
```

```python
military_df = military_df.fillna(method='ffill', axis=1)

military_df.head()
```

```python
military_long = military_df.melt(
    id_vars=['Country Name','Country Code'],
    var_name='Year',
    value_name='Military_Spending'
)

military_long.head()
```

```python
inflation_df.columns.tolist()
```

```python
keep_cols = ['Country Name','Country Code']

for year in range(2000,2025):
    keep_cols.append(str(year))

inflation_df = inflation_df[keep_cols]

inflation_df.head()
```

```python
inflation_long = inflation_df.melt(
    id_vars=['Country Name','Country Code'],
    var_name='Year',
    value_name='Inflation_Rate'
)

inflation_long.head()
```

```python
oil_df.columns
```

```python
oil_df['observation_date'] = pd.to_datetime(
    oil_df['observation_date']
)

oil_df.head()
```

```python
oil_df['Year'] = oil_df['observation_date'].dt.year

oil_df.head()
```

```python
oil_df['DCOILWTICO'] = pd.to_numeric(
    oil_df['DCOILWTICO'],
    errors='coerce'
)
```

```python
oil_yearly = oil_df.groupby('Year')['DCOILWTICO'].mean().reset_index()

oil_yearly.rename(
    columns={'DCOILWTICO':'Average_Oil_Price'},
    inplace=True
)

oil_yearly.head()
```

```python
military_long['Year'] = military_long['Year'].astype(int)

inflation_long['Year'] = inflation_long['Year'].astype(int)

conflict_df['year'] = conflict_df['year'].astype(int)
```

```python
print("Conflict Dataset :", conflict_df.shape)

print("Military Dataset :", military_long.shape)

print("Inflation Dataset :", inflation_long.shape)

print("Oil Dataset :", oil_yearly.shape)
```

```python
conflict_df.to_csv(
    'clean_conflict.csv',
    index=False
)

military_long.to_csv(
    'clean_military.csv',
    index=False
)

inflation_long.to_csv(
    'clean_inflation.csv',
    index=False
)

oil_yearly.to_csv(
    'clean_oil.csv',
    index=False
)

print("All Cleaned Files Saved")
```
